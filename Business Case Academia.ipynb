```python
# pode ser executado no google colaboratore:
# !pip install pdf_reports
import pdf_reports
import datetime as dt
```

```python
nome = input('Digite o nome do aluno(a): ')
idade = input('Digite a idade do aluno(a): -> Ex: 20 ')
peso = input('Digite o peso do aluno(a) -> Ex: 80 (Kg): ')
altura = input('Digite a altura do aluno(a) -> Ex: 1.70 (m): ')
data_avaliacao = dt.datetime.now()
```

```python
imc = float(peso) / float(altura) **2

data_avaliacao = data_avaliacao.strftime("%d_%m_%Y")
print(data_avaliacao)
print(round(imc, 2))
```

```python
status = "dentro do peso"
if imc > 25:
  status = "sobrepeso"

print(status)
```

```python
dados_aluno = {
    "nome": nome,
    "idade": idade,
    "peso": peso,
    "altura": altura,
    "imc": round(imc, 2),
    "status": status
}
```

```python
template = pdf_reports.pug_to_html("template.pug", dados=dados_aluno)
```

```python
# (use o nosso arquivo criado template, não use suas configuraçoes tradicionais. Apenas use o template.)
pdf_reports.write_report(
    template, 
    f"ficha aluno(a) {nome}_{data_avaliacao}.pdf", 
    use_default_styling=False
)
```
